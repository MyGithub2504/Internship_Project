```python
# importing necessary packages
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import WebDriverException
import selenium
import pandas as pd
import time
import warnings
import os
from time import sleep
warnings.filterwarnings("ignore")
```

```python
# for holding the resultant list
Ratings2=[]
Review2=[]
url2=[]
```

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
driver.get("https://www.amazon.in/")
```

```python
search_field=driver.find_element_by_xpath("//input[@id='twotabsearchtextbox']")
search_field.send_keys("phones")
```

```python
search_btn=driver.find_element_by_xpath("//input[@id='nav-search-submit-button']")
search_btn.click()
```

### Scrape url from Amazon

```python
try:
    for i in range(20):
        for j in driver.find_elements_by_xpath("//a[@target='_blank']"):
            url2.append(j.get_attribute("href"))
        next_button=driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
        next_button.click()
        sleep(6)
except(NoSuchElementException, WebDriverException, TimeoutException):
    pass
```

```python
print(len(url2))
```

```python
for i in url2:
    driver.get(i)
    time.sleep(3)
     #Fetching brand
    try:
        rating2=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Ratings2.append(rating2.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
     #Fetching variant
    try:
        review2=driver.find_element_by_xpath("//div[@class='a-expander-content reviewText review-text-content a-expander-partial-collapse-content']")
        Review2.append(review2.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
```

```python
Phones=pd.DataFrame({})
Phones['Ratings']=Ratings2[0:]
Phones['Reviews']=Review2[0:]
Phones
```
