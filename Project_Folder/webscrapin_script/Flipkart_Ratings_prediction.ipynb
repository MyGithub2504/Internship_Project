### Scrape url from Flipkart

```python
products=[]
try:
    for i in range(50):
        for j in driver.find_elements_by_xpath("//a[@class='_2rpwqI']"):
            url1.append(j.get_attribute("href"))
        next_button=driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
        sleep(6)
except(NoSuchElementException, WebDriverException, TimeoutException):
    pass
```

```python
for i in url1:
    driver.get(i)
    time.sleep(3)
     #Fetching brand
    try:
        rating=driver.find_element_by_xpath("//div[@class='_2d4LTz']")
        Ratings1.append(rating.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
     #Fetching variant
    try:
        review=driver.find_element_by_xpath("//div[@class='t-ZTKy']/div/div")
        Review1.append(review.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
```
