```python
# for holding the resultant list
Ratings6=[]
Review6=[]
url6=[]
```

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
driver.get("https://www.flipkart.com/")
```

```python
btn1=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
btn1.click()
```

```python
search_field6=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/form[1]/div[1]/div[1]/input[1]")
search_field6.send_keys("Monitors")
```

```python
search_btn6=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/form[1]/div[1]/button[1]")
search_btn6.click()
```

```python
products=[]
try:
    for i in range(50):
        for j in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
            url6.append(j.get_attribute("href"))
        next_button=driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
        sleep(6)
except(NoSuchElementException, WebDriverException, TimeoutException):
    pass
```

```python
for i in url6:
    driver.get(i)
    time.sleep(3)
     #Fetching brand
    try:
        rating6=driver.find_element_by_xpath("//div[@class='_2d4LTz']")
        Ratings6.append(rating6.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
     #Fetching variant
    try:
        review6=driver.find_element_by_xpath("//div[@class='t-ZTKy']/div/div")
        Review6.append(review6.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
```
