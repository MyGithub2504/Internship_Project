```python
# importing necessary packages
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import WebDriverException
import selenium
import pandas as pd
import time
import warnings
import os
from time import sleep
warnings.filterwarnings("ignore")
```

```python
# for holding the resultant list
Ratings4=[]
Review4=[]
url4=[]
```

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
driver.get("https://www.flipkart.com/")
```

```python
btn1=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
btn1.click()
```

```python
search_field4=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/form[1]/div[1]/div[1]/input[1]")
search_field4.send_keys("smart watches")
```

```python
search_btn4=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/form[1]/div[1]/button[1]")
search_btn4.click()
```

```python
try:
    for i in range(50):
        for j in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
            url4.append(j.get_attribute("href"))
        next_button=driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
        sleep(6)
except(NoSuchElementException, WebDriverException, TimeoutException):
    pass
```

```python
for i in url4:
    driver.get(i)
    time.sleep(3)
     #Fetching brand
    try:
        rating4=driver.find_element_by_xpath("//div[@class='_2d4LTz']")
        Ratings4.append(rating4.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
     #Fetching variant
    try:
        review4=driver.find_element_by_xpath("//div[@class='t-ZTKy']/div/div")
        Review4.append(review4.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
```
