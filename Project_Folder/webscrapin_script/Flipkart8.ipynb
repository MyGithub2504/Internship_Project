```python
products=[]
try:
    for i in range(50):
        for j in driver.find_elements_by_xpath("//a[@class='_2rpwqI']"):
            url8.append(j.get_attribute("href"))
        next_button=driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
        sleep(6)
except(NoSuchElementException, WebDriverException, TimeoutException):
    pass
```

```python
for i in url8:
    driver.get(i)
    time.sleep(3)
     #Fetching brand
    try:
        rating8=driver.find_element_by_xpath("//div[@class='_2d4LTz']")
        Ratings8.append(rating8.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
     #Fetching variant
    try:
        review8=driver.find_element_by_xpath("//div[@class='t-ZTKy']/div/div")
        Review8.append(review8.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
```
