### Scrape url from Amazon

```python
try:
    for i in range(50):
        for j in driver.find_elements_by_xpath("//a[@target='_blank']"):
            url3.append(j.get_attribute("href"))
        next_button=driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
        next_button.click()
        sleep(6)
except(NoSuchElementException, WebDriverException, TimeoutException):
    pass
```

```python
for i in url3:
    driver.get(i)
    time.sleep(3)
     #Fetching brand
    try:
        rating3=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Ratings3.append(rating3.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
     #Fetching variant
    try:
        review3=driver.find_element_by_xpath("//div[@class='a-expander-content reviewText review-text-content a-expander-partial-collapse-content']")
        Review3.append(review3.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
```
