```python
# importing necessary packages
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import WebDriverException
import selenium
import pandas as pd
import time
import warnings
import os
from time import sleep
warnings.filterwarnings("ignore")
```

```python
# for holding the resultant list
Ratings7=[]
Review7=[]
url7=[]
```

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
driver.get("https://www.flipkart.com/")
```

```python
btn1=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
btn1.click()
```

```python
search_field7=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/form[1]/div[1]/div[1]/input[1]")
search_field7.send_keys("Home theater")
```

```python
search_btn7=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/form[1]/div[1]/button[1]")
search_btn7.click()
```

```python
products=[]
try:
    for i in range(50):
        for j in driver.find_elements_by_xpath("//a[@class='_2rpwqI']"):
            url7.append(j.get_attribute("href"))
        next_button=driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
        sleep(6)
except(NoSuchElementException, WebDriverException, TimeoutException):
    pass
```

```python
for i in url7:
    driver.get(i)
    time.sleep(3)
     #Fetching brand
    try:
        rating7=driver.find_element_by_xpath("//div[@class='_2d4LTz']")
        Ratings7.append(rating7.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
     #Fetching variant
    try:
        review7=driver.find_element_by_xpath("//div[@class='t-ZTKy']/div/div")
        Review7.append(review7.text.replace("/n","new line"))
    except (NoSuchElementException, WebDriverException, TimeoutException):
        pass
```
